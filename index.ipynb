```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank+'*'+self.suit
    
```

```python
class Deck:
    
    def __init__(self):
        self.deck = []  # начинаем с пустого списка
        
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
                
                
    def __str__(self):
        deck_comp = ''  # начинаем с пустой строки
        for card in self.deck:
            deck_comp += '\n '+card.__str__() # добавляем строку print для каждого объекта Card
        return 'В колоде находятся:' + deck_comp

            
        
                   
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        single_card = self.deck.pop()
        return single_card
```

```python
class Chips:
    
    def __init__(self):
        self.total = 100   #  установить значение по умолчанию
        
        
    def win_bet(self,bet):
        self.total+=bet
        print(f'\nСумма выиграша: {bet}\nВаш баланс: {self.total}')
        return self.total
    
    def lose_bet(self,bet):
        self.total-=bet
        print(f'\nСумма проиграша: {bet}\nВаш баланс: {self.total}')
        if self.total==0:
            print('Деньги на вашем счете закончились!')
            while True:
                try:
                    self.total=int(input('Пополните ваш счет! Введите целое число  '))
                except:
                    print('Это не число или число не целое! Попробуйте еще раз')
                    continue
                else:
                    break
        return self.total
    
    def take_bet(self):
    
        while True:
            try:
                bet = int(input('Сделайте ставку!\nПожалуйста введите целое число:  '))
                if bet>self.total:
                    print('Ставка превышает баланс!')
                    continue

            except:
                print('Это не число или число не целое!')
                continue
            else:
                print(f'Ваша ставка: {bet}')
                break
        print('Ставка принята!')
        return bet
        
```
